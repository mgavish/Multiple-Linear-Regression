```python
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(20,3))

for xcol, ax in zip(['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view', 'condition','grade',
                    'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode', 'lat', 'long', 'sqft_living15', 
                    'sqft_lot15'], axes):
    df.plot(kind='scatter', x=xcol, y='price', ax=ax, alpha=0.4, color='b')
```

```python
continuous_var = [ 'price',  'sqft_living', 'sqft_lot', 'sqft_above','sqft_basement','yr_built', 'yr_renovated',
                  'lat', 'long', 'sqft_living15', 'sqft_lot15']
```

```python
df.hist(continuous_var,figsize = (20,20));
```

```python
for column in continuous_var:
    df[column].plot.hist(normed=True )
    df[column].plot.kde(label=column )
    plt.legend()
    plt.show()
```

```python
for column in continuous_var:
    sns.jointplot(x=column, y="price",
                  data=df, 
                  kind='reg', 
                  label=column,
                  joint_kws={'line_kws':{'color':'green'}})
    sns.regplot(df[column], df.price, label=column)
    plt.legend()
    plt.show()
```
